```python
import pandas as pd
import yfinance as yf
import numpy as np
```

```python
ticker = 'RELIANCE'
data = yf.download('^NSEBANK')
data = data.reset_index()
data['Date'] = pd.to_datetime(data['Date'])
data
```

```python
# Keeping necessary columns
data = data[['Date', 'Open', 'High', 'Low', 'Close']]
```

```python
# Calculating Daily Return Percentage
data['Return%'] = data['Close'].pct_change() * 100
data
```

```python
# Calculating Moving Averages
data['MovingAvg7'] = data['Close'].rolling(window=7).mean()
data['MovingAvg14'] = data['Close'].rolling(window=14).mean()
data['MovingAvg30'] = data['Close'].rolling(window=30).mean()

data.tail()
```

```python
# Generating signals
data['Signal'] = 0

for i in range(len(data)-2):
        # 1 corresponds to buy and -1 corresponds to sell
        if data.loc[i, 'Close'] > data.loc[i, 'MovingAvg30']:
            data.loc[i+1, 'Signal'] = 1
        elif data.loc[i, 'Close'] < data.loc[i, 'MovingAvg30']:
            data.loc[i+1, 'Signal'] = -1

data
```

### If a signal is generated today then the entry is made on opening price of the next day opening and exited if the opposite signal is given while making a position simultaneously

```python
i = 0
while i < len(data):
    j = i
    if data.loc[i, 'Signal'] == 1:
        
        # Adding the entry price and date in a new column if buy signal is given
        data.loc[i, 'Entry'] = data.loc[i, 'Open']
        data.loc[i, 'Entry Date'] = data.loc[i, 'Date']
        
        for j in range(i+1, len(data)):
            if data.loc[j, 'Signal'] == -1:
                
                # Adding the exit price and date in a new column if sell signal is given
                data.loc[i, 'Exit'] = data.loc[j, 'Open']
                data.loc[i, 'Exit Date'] = data.loc[j, 'Date']
                break
        i = j

    elif data.loc[i, 'Signal'] == -1:
        
        # Adding the entry price and date in a new column if sell signal is given
        data.loc[i, 'Entry'] = data.loc[i, 'Open']
        data.loc[i, 'Entry Date'] = data.loc[i, 'Date']
        
        for j in range(i+1, len(data)):
            if data.loc[j, 'Signal'] == 1:

                # Adding the exit price and date in a new column if buy signal is given
                data.loc[i, 'Exit'] = data.loc[j, 'Open']
                data.loc[i, 'Exit Date'] = data.loc[j, 'Date']
                break
        i = j

    elif data.loc[i, 'Signal'] == 0:
        i += 1

print(f"Entry prices:\n{data[data['Entry'].notna()]}")
print(f"\nExit price:{data[data['Exit'].notna()]}")
```

```python
# To calculate percentage return after every trade
for i in range(len(data)):

    if data.loc[i, 'Exit'] != np.nan:
        perc_ret_buy = ((data.loc[i, 'Exit'] - data.loc[i, 'Entry'])/data.loc[i, 'Entry']) * 100
        perc_ret_sell = ((data.loc[i, 'Entry'] - data.loc[i, 'Exit'])/data.loc[i, 'Entry']) * 100
        if data.loc[i, 'Signal'] == 1:
                data.loc[i, 'PercRet'] = perc_ret_buy

        elif data.loc[i, 'Signal'] == -1:
            data.loc[i, 'PercRet'] = perc_ret_sell

data
```

```python
data['Year'] = data['Date'].dt.year
annual_returns = data.groupby('Year')['PercRet'].sum()
print(f"Every Year Annual Returns:\n{annual_returns}")
print(f"Total Returns from year 2007 to 2024: {annual_returns.sum()}%")
```
